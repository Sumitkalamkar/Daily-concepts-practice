```python
# One data point
x = 2.0
y = 4.0   # true output

# Initialize parameters
w = 1.0
b = 0.0

lr = 0.1

print("Initial values")
print(f"w = {w}, b = {b}")
```

```python
y_pred = w * x + b
loss = (y_pred - y) ** 2   # MSE (single point)

print("\nForward pass")
print(f"y_pred = {y_pred}")
print(f"loss = {loss}")
```

```python
# dL/dy_pred
d_y_pred = 2 * (y_pred - y)

# dL/dw = dL/dy_pred * dy_pred/dw
d_w = d_y_pred * x

# dL/db = dL/dy_pred * dy_pred/db
d_b = d_y_pred * 1

print("\nBackpropagation gradients")
print(f"d_w = {d_w}")
print(f"d_b = {d_b}")

# Update parameters
w = w - lr * d_w
b = b - lr * d_b

print("\nUpdated parameters")
print(f"w = {w}, b = {b}")
```
