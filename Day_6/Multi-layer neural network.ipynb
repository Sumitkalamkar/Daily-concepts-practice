```python
import numpy as np 
```

```python
x = np.array([

    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y = np.array([[0], [1], [1], [0]])  # XOR output
```

```python
np.random.seed(42)

w1 = np.random.rand(2,2)
b1 = np.zeros((1,2))

w2 = np.random.rand(2,1)
b2 = np.zeros((1,1))

lr= 0.1

epoches = 5000 


```

```python
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
for epoch in range(epoches):

    # forward
    z1 = np.dot(x, w1) + b1
    a1 = tanh(z1)

    z2 = np.dot(a1, w2) + b2
    y_pred = sigmoid(z2)

    # loss
    loss = -np.mean(
        y * np.log(y_pred + 1e-8) +
        (1 - y) * np.log(1 - y_pred + 1e-8)
    )

    # backprop
    d_y = y_pred - y
    d_W2 = np.dot(a1.T, d_y)
    d_b2 = np.sum(d_y, axis=0, keepdims=True)

    d_a1 = np.dot(d_y, w2.T)
    d_z1 = d_a1 * tanh_derivative(z1)
    d_W1 = np.dot(x.T, d_z1)
    d_b1 = np.sum(d_z1, axis=0, keepdims=True)


    # Update
    w2 -= lr * d_W2
    b2 -= lr * d_b2
    w1 -= lr * d_W1
    b1 -= lr * d_b1

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```
