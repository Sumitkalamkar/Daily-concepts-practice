```python
import numpy as np 
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
def sigmoid(z):
    
    return 1/(1+np.exp(-z))

```

```python
df= load_breast_cancer()

x=df.data
y=df.target.reshape(-1,1)




```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y, test_size=0.25, random_state=42
)


```

```python
scalar= StandardScaler()

x_train=scalar.fit_transform(x_train)

x_test= scalar.transform(x_test)


```

```python
def train_adam(x,y,lr=0.01,beta1=0.9,beta2=0.999,epochs=1000):

    n,d= x.shape

    w= np.zeros((d,1))

    b=0

    mw= np.zeros_like(w)
    vw= np.zeros_like(w)

    mb=0
    vb=0

    eps=1e-8

    for t in range(1,epochs + 1):

        z=x@w+b

        y_pred = sigmoid(z)


        loss= -np.mean(y*np.log(y_pred+eps)+(1-y)*np.log(1-y_pred+eps))


        dw=(1/n)*x.T@(y_pred-y)
        db=np.mean(y_pred-y)


        mw=beta1*mw + (1-beta1)*dw
        mb=beta1*mb + (1-beta1)*db
        
        vw=beta2*vw + (1-beta2)*(dw**2)
        vb=beta2*vb + (1-beta2)*(db**2)

        mw_hat=mw/(1-beta1**t) 
        mb_hat=mb/(1-beta1**t) 
        vw_hat=vw/(1-beta2**t)
        vb_hat=vb/(1-beta2**t)


        w-=lr*mw_hat/(np.sqrt(vw_hat)+eps)
        b-=lr*mb_hat/(np.sqrt(vb_hat)+eps)


        if t%500 == 0:
            print(f"Epoch {t},Loss: {loss:4f}")

    return loss
```

```python
final_loss = train_adam(x_train, y_train)
print("Final Loss (Adam):", final_loss)
```
