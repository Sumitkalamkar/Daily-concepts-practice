```python
import numpy as np
```

```python
x = np.array([

    [0,0],
    [0,1],
    [1,0],
    [1,1]
])


y = np.array([0,1,1,1])

```

```python
weights = np.random.rand(2)
bias = np.random.rand()
lr=0.1
epoches=20



```

```python
def step_function(t):
    if t>=0:
        return 1
    return 0
```

```python
for epoch in range(epoches):
     print(f"Epoch {epoch+1}")
     for i in range(len(x)):
            z= np.dot(x[i],weights)+bias
            y_pred= step_function(z)
            error= y[i]-y_pred
            weights += lr*error*x[i]
            bias += lr*error
            print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}, Weights: {weights}, Bias: {bias}")
                     
                        
```
