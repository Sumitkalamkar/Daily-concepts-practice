```python
import numpy as np 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
df = load_breast_cancer()
x= df.data
y= df.target.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)

```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))


def train_rmsprop(x,y,lr=0.01, beta=0.9,epochs=1000):
    n,d = x.shape
    w = np.zeros((d,1))
    b=0

    sw=np.zeros_like(n)
    sb=0
    
    losses =[]
    
    for _ in range(epochs):
        z = x@w+b
        y_pred = sigmoid(z)

        eps = 1e-8

        loss = -np.mean(
            y * np.log(y_pred + eps) +
            (1 - y) * np.log(1 - y_pred + eps)
        )

        losses.append(loss)

        dw= (1/n)*x.T@(y_pred-y)
        db= np.mean(y_pred-y)
        
        sw = beta*sw + (1-beta)*dw**2
        sb = beta*sb + (1-beta)*db**2

        w -= lr*dw/(np.sqrt(sw) + 1e-8)
        b -= lr*db/(np.sqrt(sb) + 1e-8)
    
    return losses[-1]

final_loss = train_rmsprop(x_train, y_train)
print("Final Loss (RMSProp):", final_loss)
```
