```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array([1, 2, 3, 4, 5]) # Example data
y = np.array([2, 3, 5, 7, 11]) # Example data
```

```python
# True relationship: y = 2x
```

```python
# applying MSE loss function

def mse_loss(y_true,y_pred):

    return np.mean((y_true - y_pred) ** 2)


#Penalizes large errors more than small errors
#smooth and easy to optimize
```

```python
# initialize parameters:

w=0.0
b=0.0
learning_rate=0.01
epochs=60
losses=[]


```

```python
for epoch in range(epochs):

    # compute predictions
    y_pred= w*x + b

    # compute loss
    loss= mse_loss(y,y_pred)
    losses.append(loss)

    # compute gradients
    dw= -2 * np.mean(x*(y-y_pred))
    db= -2 * np.mean(y-y_pred)

    # update parameters
    w-=learning_rate* dw
    b-=learning_rate* db


    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss:.4f}, w: {w:.4f}, b: {b:.4f}")



```

```python
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.title('Loss Reduction using Gradient Descent')
plt.show()
```
