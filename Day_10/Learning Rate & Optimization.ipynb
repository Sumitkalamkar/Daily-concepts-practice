```python
import numpy as np
from  sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
data= load_breast_cancer()

x= data.data
y=data.target.reshape(-1,1) 

```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
scalar= StandardScaler()
x_train= scalar.fit_transform(x_train)  
x_test= scalar.transform(x_test)

```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def train_logistic(x,y,lr,epochs=1000):
    n,d = x.shape

    w= np.zeros((d,1))

    b=0

    losses =[]


    for _ in range(epochs):

        z=x@w + b 
        y_pred= sigmoid(z)

        loss= - (1/n) * np.sum(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
        losses.append(loss)

        dw= (1/n)* x.T @ (y_pred - y)

        db= np.mean(y_pred -y)

        w-= lr* dw 

        b-= lr* db 


    return losses





        
```

```python

learning_rates = [0.0001, 0.01,0.1,1.0]


print("Learning Rate Comaparison")




for lr in learning_rates:
    losses= train_logistic(x_train, y_train, lr)

    print(f"Learning Rate: {lr}, Final Loss: {losses[-1]:.4f}")
```
