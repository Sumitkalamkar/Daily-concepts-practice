```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

```

```python
def train_gd(x, y, lr, epochs=2000):
    n,d= x.shape
    w=np.zeros((d,1))
    b=0

    losses=[]


    for _ in range(epochs):
        z= np.dot(x,w)+b
        y_pred= sigmoid(z)

        loss= -1/n * np.sum(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))
        losses.append(loss)

        dw= 1/n * np.dot(x.T, (y_pred - y))
        db= 1/n * np.sum(y_pred - y)

        w -= lr * dw
        b -= lr * db
    return losses
```

```python
def train_momentum(X, y, lr, beta=0.9, epochs=2000):
    n, d = X.shape
    w = np.zeros((d, 1))
    b = 0

    vw = np.zeros_like(w)
    vb = 0

    losses = []

    for _ in range(epochs):
        z = X @ w + b
        y_pred = sigmoid(z)

        loss = -np.mean(
            y * np.log(y_pred + 1e-8) +
            (1 - y) * np.log(1 - y_pred + 1e-8)
        )
        losses.append(loss)

        dw = (1 / n) * X.T @ (y_pred - y)
        db = np.mean(y_pred - y)

        vw = beta * vw + dw
        vb = beta * vb + db

        w -= lr * vw
        b -= lr * vb

    return losses
```
